# Problem 1
Kullanıcıdan bir sayı alınız. Sayı 0'dan büyükse pozitif , 0'dan küçükse negatif sıfıra eşitse sıfır yazdırınız.

```python
# Kullanıcıdan bir sayı al
sayi = float(input("Bir sayı giriniz: "))

# Sayının pozitif, negatif veya sıfır olup olmadığını kontrol et
if sayi > 0:
    print("Pozitif")
elif sayi < 0:
    print("Negatif")
else:
    print("Sıfır")
```

 # Problem 2
Kullanıcıdan aldığınız bir sayının mükemmel olup olmadığını bulmaya çalışın.
Bir sayının kendi hariç bölenlerinin toplamı kendine eşitse bu sayıya "mükemmel sayı" denir. Örnek olarak, 6 mükemmel bir sayıdır. (1 + 2 + 3 = 6)


```python
def mukemmel_sayi_mi(sayi):
    # Sayının bölenlerinin toplamını bulmak için bir değişken oluştur
    toplam = 0
    
    # Sayının kendisi hariç bölenlerini bul
    for i in range(1, sayi):
        if sayi % i == 0:
            toplam += i
    
    # Bölenlerin toplamı sayıya eşitse mükemmel sayıdır
    if toplam == sayi:
        return True
    else:
        return False

# Kullanıcıdan bir sayı al
sayi = int(input("Bir sayı giriniz: "))

# Sayının mükemmel olup olmadığını kontrol et
if mukemmel_sayi_mi(sayi):
    print(f"{sayi} mükemmel bir sayıdır.")
else:
    print(f"{sayi} mükemmel bir sayı değildir.")
```

 # Problem 3
Kullanıcıdan aldığınız bir sayının "Armstrong" sayısı olup olmadığını bulmaya çalışın.
Örnek olarak, Bir sayı eğer 4 basamaklı ise ve oluşturan rakamlardan herbirinin 4. kuvvetinin toplamı( 3 basamaklı sayılar için 3.kuvveti ) o sayıya eşitse bu sayıya "Armstrong" sayısı denir.

Örnek olarak : 1634 = 1^4 + 6^4 + 3^4 + 4^4

```python
def armstrong_sayi_mi(sayi):
    # Sayıyı string olarak al ve basamak sayısını bul
    sayi_str = str(sayi)
    basamak_sayisi = len(sayi_str)
    
    # Her basamağın kuvvetini hesapla ve toplamı bul
    toplam = sum(int(rakam) ** basamak_sayisi for rakam in sayi_str)
    
    # Toplam sayı ile eşitse Armstrong sayısıdır
    if toplam == sayi:
        return True
    else:
        return False

# Kullanıcıdan bir sayı al
sayi = int(input("Bir sayı giriniz: "))

# Sayının Armstrong sayısı olup olmadığını kontrol et
if armstrong_sayi_mi(sayi):
    print(f"{sayi} bir Armstrong sayısıdır.")
else:
    print(f"{sayi} bir Armstrong sayısı değildir.")
```

İpucu: İç içe 2 tane for döngüsü kullanın. 

Aynı zamanda sayıları range() fonksiyonunu kullanarak elde edin.

```python
# Çarpım tablosunu oluşturmak için iç içe for döngüleri kullanıyoruz
for i in range(1, 11):
    for j in range(1, 11):
        # Her iki sayının çarpımını yazdırıyoruz
        print(f"{i} x {j} = {i * j}")
    # Her iç döngü tamamlandığında bir satır boşluk ekleyerek tabloyu daha okunaklı hale getiriyoruz
    print()
```

# Problem 5
Markete giren bir YetGen'li sıra beklememek için hızlı kasaya gider. Hızlı kasanın çalışma mantığı şu şekildedir:
- Kullanıcının aldığı her ürünün fiyatını tek tek girmesi gerekir.
- Kullanıcı kasadan çıkmak isterse q tuşuna basmalıdır.
- Kasa girilen fiyatları son fiyat isimli değişkene ekler.
- Değişkeni ekrana bastırır.

İpucu: while döngüsünü sonsuz koşulla başlatın ve kullanıcı q'ya basarsa döngüyü break ile sonlandırın.

```python
# Son fiyat değişkenini başlatıyoruz
son_fiyat = 0.0

print("Ürün fiyatlarını giriniz. Çıkmak için 'q' tuşuna basınız.")

# Sonsuz döngü başlatıyoruz
while True:
    # Kullanıcıdan giriş alıyoruz
    girdi = input("Ürün fiyatı: ")
    
    # Kullanıcı çıkış yapmak isterse döngüyü sonlandırıyoruz
    if girdi.lower() == 'q':
        break
    
    # Girilen fiyatı son_fiyat değişkenine ekliyoruz
    try:
        fiyat = float(girdi)
        son_fiyat += fiyat
    except ValueError:
        print("Lütfen geçerli bir fiyat giriniz.")

# Toplam fiyatı ekrana bastırıyoruz
print(f"Toplam fiyat: {son_fiyat}")
```

# Problem 6
1'den 100'e kadar olan sayılardan sadece 3'e bölünen sayıları ekrana bastırın. Bu işlemi continue ile yapmaya çalışın.

```python
# 1'den 100'e kadar olan sayılar arasında döngü oluşturuyoruz
for sayi in range(1, 101):
    # Sayı 3'e bölünmüyorsa döngünün sonraki adımına geçiyoruz
    if sayi % 3 != 0:
        continue
    # Sayı 3'e bölünüyorsa ekrana bastırıyoruz
    print(sayi)
```

# Problem 7
Bir YetGen Core Python mezunu MEF Üniversitesi'nde Bilgisayar Bilimleri Doktora programının mülakatına girmeye hak kazanıyor. Mülakata hazırlanırken internetten soru araştırıyor. Karşısına çıkan ünlü FizzBuzz sorusuna benzer YetGen sorusunda kendisini yeterince hazır hissetmiyor. Bu soruyu çözmesine destek olmak ister misin?

Soru aşağıdaki gibidir:

1’den 100’e kadar sayıları yazdırıyoruz. Ancak eğer sayı 3’e bölünüyor ise Yet, sayı 5’e bölünüyorsa Gen, hem 3’e hem 5’e bölünüyorsa YetGen yazdırıyoruz.

```python
# 1'den 100'e kadar sayılar arasında döngü oluşturuyoruz
for sayi in range(1, 101):
    # Hem 3'e hem de 5'e bölünüyorsa "YetGen" yazdırıyoruz
    if sayi % 3 == 0 and sayi % 5 == 0:
        print("YetGen")
    # Sadece 3'e bölünüyorsa "Yet" yazdırıyoruz
    elif sayi % 3 == 0:
        print("Yet")
    # Sadece 5'e bölünüyorsa "Gen" yazdırıyoruz
    elif sayi % 5 == 0:
        print("Gen")
    # Hiçbirine bölünmüyorsa sayının kendisini yazdırıyoruz
    else:
        print(sayi)
```
